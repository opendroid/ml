# Example: Sobel filter

```python
from scipy.signal import convolve2d
import numpy as np
import matplotlib.pyplot as plt
from skimage import color, data
```

```python
astronaut = color.rgb2gray(data.astronaut())  # Sample grayscale image

plt.figure(figsize=(2, 2))
plt.imshow(astronaut)
plt.axis('off')
plt.show()
```

```python
# Sobel filter
sobel_vertical = np.array([[-1, 0, 1],
                           [-2, 0, 2],
                           [-1, 0, 1]])
```

```python
edge_vertical = convolve2d(astronaut, sobel_vertical,
                           mode='same', boundary='symm')
```

```python
# Show the images
plt.figure(figsize=(2, 2))
plt.imshow(edge_vertical, cmap='gray')
plt.axis('off')
plt.show()
```
